# Importing libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score 
import pickle
```

# Dataset Loading

```python
df=pd.read_csv('dataset/diabetes.csv')
```

# Dataset Info

```python
df
```

```python
print(f"Row x Col. : {df.shape}")
```

```python
df.info()
```

# Data Preprocessing

```python
df.isna().sum()
```

```python
df.head()
```

```python
df['Age'].unique()
```

```python
col_with_zero = {'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI'}
```

```python
import numpy as np 
```

```python
for col in col_with_zero:
    df[col] = df[col].replace(0, np.nan)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
for col in col_with_zero:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
df.isna().sum()
```

# Train & Test Splitting

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('Outcome', axis = 1)
y = df['Outcome']
```

```python
print(x.columns)
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2,  random_state= 42)
```

```python
print(f"x shape : {x.shape}, x_train.shape : {x_train.shape}, x_test.shape : {x_test.shape}")
```

# Model Training (Logistic Regession)

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

# Model accuracy score

```python
from sklearn.metrics import accuracy_score 
```

```python
y_pred = model.predict(x_test)
```

```python
acc = accuracy_score(y_test, y_pred)
```

```python
print(f"accuracy_score = {acc}")
```
